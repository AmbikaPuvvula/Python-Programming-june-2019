### Day Objectives
- App Store Dataset
    - Category with highest average user rating for paid apps
    - Most frequent Price point > 0
    - Compare average user rating for paid vs free gaming apps

```python
filepath = 'DataFiles/AppleStore.csv'
import pandas as pd
def readCSVdata(filepath):
    return pd.read_csv(filepath)
readCSVdata(filepath)
```

```python
# Most frequent Price point > 0
appdata = readCSVdata(filepath)
def highFrequentPrice(f):
    return list(f.price.value_counts().index)[1]
highFrequentPrice(appdata)
```

```python
# Average user rating for paid apps
appdata = readCSVdata(filepath)
l=[]

def paidApps(f):
   c=0
   for i in range(len(f.values[0::])):
       column = f.values[i][5]
       if column > 0.00:
           l.append(f.values[i][8])
           c+=1
   #print(sum(l))      
   return sum(l)/c
paidApps(appdata)
```

```python
#Compare average user rating for paid vs free gaming apps
appdata = readCSVdata(filepath)
def compare(appdata):
    p=paidApps(appdata)
    k=appdata.user_rating.values
    count=l=0
    for i in range(len(k)):
        if appdata['price'][i]==0 and appdata['prime_genre'][i]=='Games':
            count+=k[i]
            l+=1
    if (count/l)>p:
        print('Average user_rating free gaming apps')
    else:
        print('Average user_raring paid apps')
compare(appdata)
```

```python
# Linear graph
import matplotlib.pyplot as plt
plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time
distance = list(range(2,11,2))
time = list(range(1,6))
#distance
#time
# d = 2 * t
plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
# Non linear graph
# v = a t
t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i] * i for i in range(len(t))]
plt.plot(t,v,'go')
plt.grid()
plt.show()
```

```python
# Graph for numbers vs Squares(or Cubes)
n = list(range(10))
s = [i ** 100 for i in n]
plt.plot(n,s)
plt.grid()
plt.show()
```

```python
# Sinusoidal Graph 
# Voltage vs Time
import numpy as np
t = [i/10 for i in range(1,21)]
v = [1 + np.sin(2 * np.pi * i) for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v,'go')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# plot income of Arkansas from 2005-2013
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
years = df.columns[2:]
incomeArkansas = df.values[3,2:]
#years
#incomeArkansas
plt.title('Income of Arkansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.grid()
plt.savefig('Images/IncomeArkansas.png')
plt.show()
```
